```python
import seaborn as sns
df1_3 =sns.load_dataset("mpg")
sns.countplot(x='cylinders', data=df1_3)
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.title('Distribution of Cylinders')
plt.show()
```

```python
cylinders_counts = df1_3['cylinders'].value_counts()

plt.pie(cylinders_counts, labels=cylinders_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Pie Chart of Cylinders')
plt.show()
```

```python
sns.histplot(data=df1_3, x='mpg', bins=20, kde=True)
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Frequency')
plt.title('Histogram of mpg Variable')
plt.grid(True)
plt.show()
```
