```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


subsidies_series = pd.Series([856, 560, 395, 311, 285, 203, 199, 179, 168, 157, 147, 142, 132, 124, 117])

sorted_subsidies = subsidies_series.sort_values()


cumulative_subsidies = sorted_subsidies.cumsum()

cum_percentage_subsidies = cumulative_subsidies / cumulative_subsidies.max()
population_percentage = np.linspace(0, 1, len(sorted_subsidies))

plt.figure(figsize=(8, 6))
plt.plot(population_percentage, cum_percentage_subsidies, marker='o', linestyle='-', color='b', label='Lorenz Curve')


plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Line of Equality')

plt.title('Lorenz Curve of Subsidies Distribution')
plt.xlabel('Cumulative Population Percentage')
plt.ylabel('Cumulative Subsidies Percentage')
plt.legend()
plt.grid(True)
plt.show()
```
