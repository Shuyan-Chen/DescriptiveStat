```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


subsidies_series = pd.Series([856, 560, 395, 311, 285, 203, 199, 179, 168, 157, 147, 142, 132, 124, 117])

sorted_subsidies = subsidies_series.sort_values()


cumulative_subsidies = sorted_subsidies.cumsum()

cum_percentage_subsidies = cumulative_subsidies / cumulative_subsidies.max()
population_percentage = np.linspace(0, 1, len(sorted_subsidies))

plt.figure(figsize=(8, 6))
plt.plot(population_percentage, cum_percentage_subsidies, marker='o', linestyle='-', color='b', label='Lorenz Curve')


plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Line of Equality')

plt.title('Lorenz Curve of Subsidies Distribution')
plt.xlabel('Cumulative Population Percentage')
plt.ylabel('Cumulative Subsidies Percentage')
plt.legend()
plt.grid(True)
plt.show()
```

```python
income = pd.read_excel('D:\LabHw\DescriptiveStat\data\incomeRF_data.xlsx')
income.columns
```

```python
LU_income = income.loc[income['province'] == "La Union", 'income']
```

```python
LU_sorted = LU_income.sort_values()
lorenz_curve = LU_sorted.cumsum()/LU_sorted.sum()
y = pd.concat([pd.Series([0]), lorenz_curve])
x = pd.concat([pd.Series(np.arange(0,1,1/len(LU_income))), pd.Series([1])])

plt.plot(x,y, label="lorenz curve")
plt.plot([0,1],[0,1],linestyle = '--', color = 'gray', label='Line of equality')
plt.xlabel('fraction of population; from the poorest to richest')
plt.ylabel('Cumulative share of income')
plt.title('Lorenz curve: La Union')
plt.legend
plt.show()




plt.subplot(1,2,1)
PS_income = income.loc[income['province'] == "Pangasinan", 'income']
PS_sorted = PS_income.sort_values()
lorenz_curve = PS_sorted.cumsum()/PS_sorted.sum()
y = pd.concat([pd.Series([0]), lorenz_curve])
x = pd.concat([pd.Series(np.arange(0,1,1/len(PS_income))), pd.Series([1])])

plt.plot(x,y, label="lorenz curve")
plt.plot([0,1],[0,1],linestyle = '--', color = 'gray', label='Line of equality')
plt.xlabel('fraction of population; from the poorest to richest')
plt.ylabel('Cumulative share of income')
plt.title('Lorenz curve: Pangasinan')
plt.legend
plt.show()
```
