```python
import numpy as np
import pandas as pd
dataset1 = pd.Series([1, 2, 3, 4, 5, 6, 6, 7, 8, 9, 10, 11])
dataset2 = pd.Series([4, 5, 5, 5, 6, 6, 6, 6, 7, 7, 7, 8])
dataset1.mean()
dataset2.mean()
dataset1.median()
dataset2.median()
dataset1.mode()
dataset2.mode()
dataset1.quantile(q=0.25)
dataset2.quantile(q=0.25)
dataset1.quantile(q=0.75)
dataset2.quantile(q=0.75)
```

```python
dataset1.describe()
dataset2.describe()
```

```python
freq1 = dataset1.value_counts(normalize = True)
freq1
freq2 = dataset2.value_counts(normalize = True)
freq2
```

```python
import matplotlib.pyplot as plt
plt.stem(freq2.index, freq2)
```

```python
plt.stem(freq1.index, freq1)
```
